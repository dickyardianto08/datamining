## Tugas UAS | Jakarta Air Polution

```python
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score, confusion_matrix,accuracy_score, classification_report
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dickyardianto08/dataset/main/jakarta_air_polution.csv')
df.head()
```

```python
df = df[df.pm10 != "---"]
df = df[df.so2 != "---"]
df = df[df.o3 != "---"]
df = df[df.co != "---"]
df = df[df.no2 != "---"]
df = df[df.critical != "---"]
df = df[df.categori != "---"]
df.dropna(subset = ["critical"], inplace=True)
df.dropna(subset = ["max"], inplace=True)
df= df[df['max'] != 0]
```

```python
df.isnull().sum()
```

```python
df
```

```python
df= df.drop(['stasiun','tanggal','critical','max'], axis=1)
```

```python
df
```

```python
from sklearn.metrics import make_scorer, accuracy_score,precision_score
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold,train_test_split,cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
import xgboost as xgb

from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import f1_score
```

```python
X=df.iloc[:,0:5].values
y=df.iloc[:,5].values
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
import pickle
model = RandomForestClassifier()
model.fit(X_train, y_train)
# save the model to disk
filename = 'jakarta_polution.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
models = [GaussianNB(), RandomForestClassifier(), KNeighborsClassifier()]
scores = dict()

for m in models:
    m.fit(X_train, y_train)
    y_pred = m.predict(X_test)

    print(f'model: {str(m)}')
    print(classification_report(y_test,y_pred, zero_division=1))
    print('-'*30, '\n')
```
